```python
import pandas as pd
import numpy as np
import os
```

```python
data_path = os.path.join('..', 'data', 'pokemon_egg_group.csv')
dtype_dict = {
    'pokemon_id': 'int64',
    'cycles': 'int64',
    'steps_gen_ii': 'Int64',
    'steps_gen_iii': 'Int64',
    'steps_gen_iv': 'Int64',
    'steps_gen_v-vi': 'Int64',
    'steps_gen_vii': 'Int64',
    'steps_gen_viii-ix': 'Int64'
}
df = pd.read_csv(data_path, dtype=dtype_dict)
```

```python
df
```

```python
# Get the egg groups for Totodile and Weavile, excluding NaN values
totodile_egg = set(df.loc[df['pokemon'] == 'Totodile', ['egg_group_1', 'egg_group_2']].values.flatten())
totodile_egg = {group for group in totodile_egg if pd.notna(group)}

weavile_egg = set(df.loc[df['pokemon'] == 'Weavile', ['egg_group_1', 'egg_group_2']].values.flatten())
weavile_egg = {group for group in weavile_egg if pd.notna(group)}

# Filter the DataFrame based on the egg groups
filtered_df = df[df.apply(lambda row: 
                          bool((set([row['egg_group_1'], row['egg_group_2']]) & totodile_egg)) &
                          bool((set([row['egg_group_1'], row['egg_group_2']]) & weavile_egg))
         , axis=1)]
```
